```python
import uuid

ids = [str(uuid.uuid5(uuid.NAMESPACE_DNS, machine)) for machine in machines]
unique_ids = list(set(ids))
unique_ids
```

```python
# Ensure that only docs that correspond to unique ids are kept and that only one of the duplicate ids is kept
seen_ids = set()
unique_docs = [machine for machine, id in zip(machines, ids) if id not in seen_ids and (seen_ids.add(id) or True)]
unique_docs
```

```python
existing_ids = machine_collection.get()["ids"]

# Filter out the unique_ids and unique_docs if the unique_id is already in existing_ids
filtered_pairs = [(uid, doc) for uid, doc in zip(unique_ids, unique_docs) if uid not in existing_ids]

if filtered_pairs:
    unique_ids, unique_docs = zip(*filtered_pairs)
else:
    unique_ids, unique_docs = [], []
```

```python
unique_docs, unique_ids
```

```python
machine_collection.add(
    documents=unique_docs,
    ids=unique_ids,
)
```
