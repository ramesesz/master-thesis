```python
from os.path import expanduser
from prompts import *
from langchain.chains import LLMChain
from langchain.memory import ConversationBufferMemory
from langchain_experimental.chat_models import Llama2Chat
from langchain.prompts.chat import (
    ChatPromptTemplate,
    HumanMessagePromptTemplate,
    MessagesPlaceholder,
)
from langchain_core.messages import SystemMessage
from langchain_community.llms import LlamaCpp
```

```python
template_messages = [
    SystemMessage(content="You are a helpful assistant."),
    MessagesPlaceholder(variable_name="chat_history"),
    HumanMessagePromptTemplate.from_template("{text}"),
]
prompt_template = ChatPromptTemplate.from_messages(template_messages)
```

```python
model_path = expanduser("../models/llama-2-7b-chat.Q5_K_S.gguf")

llm = LlamaCpp(
    model_path=model_path,
    streaming=True,
)
model = Llama2Chat(llm=llm)
```

```python
memory = ConversationBufferMemory(memory_key="chat_history", return_messages=True)
chain = LLMChain(llm=model, prompt=prompt_template, memory=memory)
```

```python
chain.invoke(
    # input="What can I see in Vienna? Propose a few locations. Names only, no details"
    input="""
    Translate to english:
Jeder Zugriff auf DLR Server bedeutet eine Ausfuhr von Daten. Dieser Zugriff muss exportkontrollrechtliche geprüft werden.
DLR Hardware kann der US Exportkontrolle unterliegen und die Ausfuhr für bestimmte Länder nicht möglich sein
Bei geheimschutzrelevanten Themen ist eine Ausnahmegenehmigung des/der Sicherheitsbevollmächtigten
notwendig. Dies Ausnahmen werden in der Regel nicht erteilt, wenn außerhalb der NATO Länder gereist
wird.
Folgende Maßnahmen werden bei Tätigkeiten in Drittstaaten empfohlen:
 Bei Reisen ins Ausland insbesondere in Drittsaaten muss bei Mitnahme von DLR Informationen eine
Exportkontrollprüfung durchgeführt werden. Es dürfen keine Dual-Use oder Rüstungsgüter (Waren,
Software und Technologie) ohne Genehmigung in ein Drittland ausgeführt werden. Hier finden die
internen Regeln der Exportkontrolle Anwendung.
 Keine vertraulichen Gespräche am Handy
 Vorsicht bei Arbeit an öffentlichen Plätzen oder im Verkehrsmittel
 Blickschutzfolien
 Achtung Zuhörer
 Starke Passwort Nutzung
 Merkblatt Dienstreisen ins Ausland
 Dienstreiselaptop
 So wenig Daten, wie möglich
 Die wichtigsten Sachen immer bei sich führen
 Mobile Geräte und wichtige Unterlagen nicht unbeaufsichtigt lassen
 Vorsicht bei neuen Kontakten bzw. Gesprächspartnern (auch im privaten Rahmen)
 Keine USB-Sticks in fremde Rechner – keine fremde USB-Sticks in eigene Rechner
 Nur E-Mails aus vertrauenswürdigen Quellen öffnen (Vorsicht bei Anhängen!!!)
    """
)
```
